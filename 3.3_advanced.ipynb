```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# All avaiable Styles
print(plt.style.available)
```

```python
# Setting Styles
plt.style.use('ggplot')

# plt.xkcd() # Setting a comic Style

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()

plt.show()
```

```python
# Filing Line Plots
plt.style.use('ggplot')

python = np.array([40, 52, 58, 73, 102, 105, 121])
javascript = np.array([35, 40, 50, 100, 78, 68, 82])

experience = [0, 1, 2, 3, 5, 10, 20]

plt.plot(experience, python, 'g', label='Python Dev')
plt.plot(experience, javascript, 'r', label='Javascript Dev')
plt.title('Python & Javascript Salary Curve')
plt.xlabel('Experience (Year)')
plt.ylabel('Salary (per Year)')

plt.fill_between(experience, python, javascript, where=(python > javascript), interpolate=True, alpha=0.25, color='green',label='Avobe' )
plt.fill_between(experience, python, javascript, where=(python <= javascript), interpolate=True, alpha=0.25, color='red',label='Below' )
plt.legend(loc=(0.65, 0.07)) 

plt.tight_layout()

plt.show()
```

## Subplot

```python
plt.style.use('seaborn')

# Figure : A Container of multiple plots
# Axis : A plot
# Use subplot when working with multiple figure

x = np.array([1, 2, 3, 4, 5, 6])
channel_1 = np.array([100, 150, 260, 320, 880, 1000])
channel_2 = np.array([100, 125, 280, 430, 520, 1000])

fig, ax = plt.subplots() # default : 1 by 1 ... only one plot
ax.plot(x, channel_1, label='Channel 1')
ax.plot(x, channel_2, label='Channel 2')

ax.set_xlabel('Days')
ax.set_ylabel('Subcriptions')
ax.set_title("Channel 1 vs 2")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Display multiple plots in single Figure
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(x, channel_1, label='Channel 1')
ax1.plot(x, channel_2, label='Channel 2')
ax1.set_ylabel('Subcriptions')
ax1.set_title("Channel 1 vs 2")
ax1.legend()

ax2.plot(x, np.cumsum(channel_1), label='Channel 1')
ax2.plot(x, np.cumsum(channel_2), label='Channel 2')
ax2.set_xlabel('Days')
ax2.set_ylabel('Cumulative Subcriptions')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Display in different Figures
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(x, channel_1, label='Channel 1')
ax1.plot(x, channel_2, label='Channel 2')
ax1.set_ylabel('Subcriptions')
ax1.set_xlabel('Days')
ax1.set_title("Channel 1 vs 2")
ax1.legend()

ax2.plot(x, np.cumsum(channel_1), label='Channel 1')
ax2.plot(x, np.cumsum(channel_2), label='Channel 2')
ax2.set_xlabel('Days')
ax2.set_ylabel('Cumulative Subcriptions')
ax2.set_title("Channel 1 vs 2")
ax2.legend()

plt.tight_layout()
plt.show()
```
