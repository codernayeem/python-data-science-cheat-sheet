```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

## Simple Poltting using PyPlot

```python
# Simple plotting
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# ploting with only ypoints (default xpoints=[0, 1, 2, 3, ...]) 
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints)
plt.show()
```

```python
# Multiple Points with label, title, legend
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend() # this shows the label
# plt.legend(['linear', 'quadratic', 'cubic']) # This could be done by this way too
plt.tight_layout() # this gives some paddings...

plt.savefig('data\\plot1.png') # this will save this figure as png format

plt.show()
```

```python
# Multiple plots in a figure with label, title

# The subplots() function takes three arguments that describes the layout of the figure.
# The layout is organized in rows and columns, which are represented by the first and second argument.
# The third argument represents the index of the current plot.

#plot 1:
plt.subplot(2, 2, 1)
plt.plot([3, 8, 1, 10])

#plot 2:
plt.subplot(2, 2, 2)
plt.plot([7, 6, 9, 8])

#plot 2:
plt.subplot(2, 2, 3)
plt.plot([2, 9, 7, 2])

#plot 2:
plt.subplot(2, 2, 4)
plt.plot([3, 8, 9, 3])

plt.show()
```

## Markers

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o') # no line (only markars)
plt.show()
```

```python
values1 = np.array([3, 8, 1, 9])
values2 = np.array([8, 7, 9, 6])
values3 = np.array([5, 1, 2, 8])
x = [1, 2, 3, 4]

# arguments releated markar >
# marker : markar style
# ms / markersize : Markar Size
# mec / markeredgecolor : Marker Edge Color
# mfc / markerfacecolor  : Marker Face Color

plt.plot(x, values1, marker='o', ms=10, mec='r', mfc='y', label='Line 1')
plt.plot(x, values2, marker='o', mec='red', mfc='r', label='Line 2')
plt.plot(x, values3, marker='o', markersize=12, mfc='#9C0550', label='Line 3')

plt.legend(loc='lower right', fontsize="large", shadow=True)
plt.show()
```

## Line

```python
values1 = np.array([3, 8, 1, 9])
values2 = np.array([8, 7, 6, 7])
values3 = np.array([5, 1, 5, 8])
x = [1, 2, 3, 4]

# arguments releated markar >
# ls / linestyle : Line style
# color : Line Color
# lw / linewidth : Line Width
# mfc / markerfacecolor  : Marker Face Color

plt.plot(x, values1, ls='dotted', color='red', lw=5)
plt.plot(x, values2, ls=':', color='#9C0550')
plt.plot(x, values3, ls='-', color='y', lw=3)
plt.show()
```

```python
# Draw two lines by specifiyng the x- and y-point values for both lines:
    
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

## Label & Title

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y, label="Data 1")
plt.plot(x, y*1.5, label="Data 2")
plt.legend(loc='upper right')
plt.show()
```

## Set Markar+Line+Color String Format

```python
ypoints = np.array([3, 8, 2, 10])

plt.plot(ypoints, 'o:r') # marker(o)+line(:)+color(r)
plt.plot(ypoints*2, 'xg') # marker(o)+color(r)
plt.plot(ypoints*3, '--k') # line(--)+color(k)
plt.show()
```

## Scatter

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])

# s / size : dot size
# c / color : dot color
# alpha : dots opacity

plt.scatter(x, y1, s=100, c='yellow', edgecolor='green', lw=2, alpha=0.5)
plt.scatter(x, y2, c='red')
plt.show()
```

```python
# give each dots it's own color and size
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.figure(figsize=(10, 6)) # set figure size
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
# cmap : colormap is used to get colors form colors array
# there are many more colormap like: gray, greens, copper etc.

plt.colorbar() # gives the side color bar

plt.show()
```

## Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 4, 10])

plt.bar(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 4, 10])

plt.barh(x,y) # horizontal bar
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 4, 10])

# color : bar color
# width : the width of each bar (vertical bar)
# height : the width of each bar (horizontal bar)

plt.bar(x, y, color='r', width=0.2)
plt.show()
```

```python
# Display more that one bars

# better -> use seaborn

y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])
y3 = np.array([2, 5, 8, 10])

years = ['1938', '1940', '1942', '1944']

x_index = np.arange(len(y1))

width = 0.25

plt.xticks(ticks=x_index, labels=years)

plt.bar(x_index-width, y1, width=width)
plt.bar(x_index, y2, width=width)
plt.bar(x_index+width, y3, width=width)

plt.show()
```

## Histograms

```python
x = np.random.normal(170, 10, 250)

bins = [170, 180, 190, 200, 210]

plt.hist(x, edgecolor='k')
plt.show()
```

```python
x = np.random.randint(170, 250, size=(150))

plt.hist(x, edgecolor='k', bins=8, log=True)
plt.show()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

# 80-100: Normal
# 100-125: Pre-diabetic
# 125-150: Diabetic

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
plt.show()
```

```python
# Mutiple data samples in a histogram
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'])
plt.legend()
plt.show()
```

```python
# histtype
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, histtype='step')
plt.show()
```

```python
# horizontal hist
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
plt.show()
```

## Pie-Chart

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
labels = ['Banana', 'Mango', 'Lichi', 'Orange']
colors = ['yellow', 'green', 'red', 'orange']

plt.pie(y, labels=labels, colors=colors, counterclock=False, startangle=90, radius=1.5)
plt.legend(loc='lower right')
plt.show()
```

```python
y = np.array([35, 25, 25, 15, 30, 20])
mylabels = ["Apples", "Bananas", "Cherries", "Dates", "Lichi", "Guava"]
myexplode = [0.2, 0, 0, 0, 0, 0]

plt.pie(y, labels=mylabels, explode=myexplode, wedgeprops={'edgecolor': 'black'}, shadow=True, autopct='%1.1f%%')
plt.show() 
```

## Stack Plots

```python
minutes = np.arange(1, 10)

player1 = [1, 2, 3, 3, 4, 4, 4, 4, 5]
player2 = [1, 1, 1, 1, 2, 3, 3, 4, 4]
player3 = [1, 1, 1, 2, 2, 3, 3, 3, 3]

labels = ["Sakib", 'Tamim', 'Sabbir']
colors = ['r', 'g', 'orange']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)
plt.legend(loc='upper left')

plt.show()
```

## Time Series

```python
from datetime import datetime
from matplotlib import dates as mpl_dates
plt.style.use('ggplot')

dates = [
    datetime(2020, 9, 10),
    datetime(2020, 9, 11),
    datetime(2020, 9, 12),
    datetime(2020, 9, 13),
    datetime(2020, 9, 14),
    datetime(2020, 9, 15),
    datetime(2020, 9, 16),
]
work_hour = [8, 7, 7.5, 5, 6, 8, 4.5]

plt.plot_date(dates, work_hour, 'o:r')

plt.gcf().autofmt_xdate()
date_fmt = mpl_dates.DateFormatter('%b %d %Y')
plt.gca().xaxis.set_major_formatter(date_fmt)

plt.title('Working Hour')
plt.xlabel('Dates')
plt.ylabel('Work Time')

plt.tight_layout()
plt.show()
```

## Passing data as dict

```python

data = {
    'Days' : np.array([1, 2, 3, 4]),
    'Bangladesh' : np.array([4, 8, 16, 18]),
    'Combodia'   : np.array([3, 5, 12, 17]),
}

plt.plot('Days', 'Bangladesh', label='Bangladesh', c='k', data=data)
plt.plot('Days', 'Combodia', label='Combodia', data=data)
plt.title('New Baby Birth')
plt.xlabel('Days')
plt.ylabel('Babies')
plt.xticks(ticks=np.arange(1, 5), labels=['First', 'Second', 'Third', 'Fourth'])
plt.yticks(ticks=np.arange(1, 20))
plt.legend()
plt.show()
```
